```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
wp_26 = np.load('waypoints/26_agus_wp_lurus.npy')
# wp_26 = np.load('waypoints/26_agus_wp_belok.npy')
# wp_26 = np.load('waypoints/26_agus_wp_S.npy')
```

```python
plt.plot(wp_26[:,0], wp_26[:,1], label='26 Agustus 2020')
plt.legend()
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.title('')
plt.show()
```

```python
# Initial Position at a specific point
x0 = -96.02
y0 = 203.88
temp = np.load('waypoints/26_agus_wp_lurus.npy')

# Shift to the new Position
wp_new = np.copy(wp_26)
wp_new[:,0] = wp_new[:,0] - temp[0,0] + x0
wp_new[:,1] = wp_new[:,1] - temp[0,1] + y0

# Align the initial position
plt.plot(wp_26[:,0], wp_26[:,1], label='26 Agustus 2020')
plt.plot(wp_new[:,0], wp_new[:,1], label='27 Agustus 2020')
plt.legend()
plt.xlabel("X (m)")
plt.ylabel("Y (m)")
plt.show()
```
